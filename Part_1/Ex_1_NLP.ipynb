### Loading the subtitles files.

```python
import pandas as pd
```

```python
eng_sub = open('tep.lowercased.en','r', encoding="utf8")
fa_sub = open('tep.lowercased.fa','r', encoding="utf8")
eng_lines = eng_sub.readlines()
fa_lines = fa_sub.readlines()

subfile_lines=[]
count=0
for eng_line in eng_lines:
    subfile_lines.append({'eng': eng_line[0:-2], 'fa': fa_lines[count][0:-2]})
    count+=1

subfile_lines
```

### Tokenizing the sub_files and establishing the base dictionary.

```python
import re
import nltk
import hazm as hzm
from nltk.corpus import stopwords

nltk.download('stopwords')
stop_words = set(stopwords.words('english'))
frequency_dictionary={}
normalizer = hzm.Normalizer()
k=1
temp_en= {}
temp_fa= {}
cnt_En= 0 
cnt_Fa = 0

for subs in subfile_lines:
    
#     if(i%5000 == 0):
#         print(f'{i}/{len(subfile_lines)}')

    normalized_fa = normalizer.normalize(subs['fa'])
    fa_words = hzm.word_tokenize(normalized_fa)
    for idx in fa_words:
        if idx in temp_fa.keys():
            temp_fa[idx]+=1
        else:
            temp_fa[idx]=1
        cnt_Fa +=1
        

    # Convert all to lower
    lower_string = subs['eng'].lower()
    # remove numbers
    no_num = re.sub(r'\d+','',lower_string)
    # remove punctuations except words and space 
    no_punc = re.sub(r'[^\w\s]','', no_num)
    # remove white spaces
    no_wspace = no_punc.strip()
    # Convert string to list of words
    eng_words = [no_wspace][0].split()
    
    # remove stopwords
    no_stopwords_str = []
    for i in eng_words:
        if not i in stop_words:
            no_stopwords_str.append(i)
            if i in temp_en.keys():
                temp_en[i]+=1
            else:
                temp_en[i]=1
            cnt_En +=1
            
            
    for eng_word in no_stopwords_str:
        for fa_word in fa_words:
            if eng_word+", "+fa_word in frequency_dictionary.keys():
                
#             founded = False
#             for dictionary in frequency_dictionary:
#                 if dictionary['eng'] == eng_word and dictionary['fa'] == fa_word:
#                     dictionary['count']+=1

                    frequency_dictionary[eng_word+", "+fa_word]+=1

    
#                     founded = True
#                     break

            else:
                frequency_dictionary[eng_word+", "+fa_word]=1

    k=k+1
    if(k%100000 == 0):
        print(f'{k}/{len(subfile_lines)}')

frequency_dictionary
```

# Designing FrontEnd

```python
from tkinter import *
window = Tk()
window.title("Statical English-Farsi Translator")

# ===================== Labels ========================

l1 = Label(window, text="English Entry")
l1.grid(row=0, column=0)

l2 = Label(window, text="High possibility Farsi Translation")
l2.grid(row=1, column=0)

l2 = Label(window, text="Low possibility Farsi Translation")
l2.grid(row=2, column=0)

e1 = Entry(window, width=100)
e1.grid(row=0, column=1)

b1 = Button(window, text="Translate", width=12, height=3, command=lambda: translator())
b1.grid(row=0, column=4)

b2 = Button(window, text="Clear", width=12, height=3, command=lambda: clear())
b2.grid(row=1, column=4)
 
list1 = Listbox(window, width=100, height=3)
list1.grid(row=1, column=1)

list2 = Listbox(window, width=100, height=6)
list2.grid(row=2, column=1)


def translator():
    i = e1.get()
    
    tops = temp.loc[{i}].sort_values(by='static_Accuracy',ascending=False)
    tops = tops.reset_index()
    downs = tops[tops['static_Accuracy']<0.3][:3]
    tops = tops[tops['static_Accuracy']>=0.3][:3]
    
    t1 = []
    t2 = []
    t1.append(tops[['Farsi','static_Accuracy']] )
    t2.append(downs[['Farsi','static_Accuracy']] )
    
    print(t1)
    print(type(t1))
    
    list1.insert(1,t1)
    list2.insert(2,t2)
    
    return t1, t2


def clear():
    list1.delete(0, END)
    list2.delete(0, END)


window.mainloop()
```
